```python
N, K = map(int, input().split())
que = list(map(int, input().split()))
bag = []

def operate(que, bag):
    #print(que, bag)
    retlist = [[que, bag]]
    if len(que) > 0:
        retlist.append([que[1:], bag + [que[0]]])
        retlist.append([que[:-1], bag + [que[-1]]])
    if len(bag) > 0:
        min_bag = min(bag)
        bag.pop(bag.index(min_bag))
        retlist.append([[min_bag] + que, bag])
        retlist.append([que + [min_bag], bag])
    print(retlist)
    return retlist

all_possible = [[que, bag]]
for _ in range(K):
    for possible in all_possible[:]:
        all_possible +=  operate(possible[0], possible[1])
        #all_possible.append(operate(possible[0], possible[1]))
    print(all_possible)
    print()

#print(all_possible)
values = [sum(x[1]) for x in all_possible]
print(max(values), all_possible[values.index(max(values))])
# 
```

```python
def main():
    N = int(input())
    A = list(map(int, input().split()))
    min_A = min(A)
    min_A_idx = A.index(min_A)
    A = A[min_A_idx:] + A[:min_A_idx]
    
    for s in range(A[0] * 2, -1, -2):
        mt = [None] * N
        mt[0] = s
        for n in range(N-1):
            if mt[n] > A[n+1] * 2:
                break
            mt[n+1] = (A[n+1] - mt[n]//2) * 2

        if mt[N-1] != None:
            if mt[N-1] //2 + mt[0] // 2 == A[0]:
                if min_A_idx == 0:
                    print(mt[-1], ' '.join(list(map(str, mt[:-1]))))
                    return
                print(' '.join(list(map(str, mt[N-min_A_idx-1:]))), ' '.join(list(map(str, mt[:N-min_A_idx-1]))))

if __name__ == '__main__':
    main()
```
